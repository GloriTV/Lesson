```python
import pandas as pd
import category_encoders as ce   #  category_encoders
```

```python
# инициализируем информацию об одежде
clothing_list = [
    ['xxs', 'dress'],
    ['xxs', 'skirt'],
    ['xs', 'dress'],
    ['s', 'skirt'],
    ['m', 'dress'],
    ['l', 'shirt'],
    ['s', 'coat'],
    ['m', 'coat'],
    ['xxl', 'shirt'],
    ['l', 'dress']
]

clothing_itol = pd.DataFrame(clothing_list, columns = ['size',  'type'])
clothing_itol
```

Метод fit_transform

```python
# импортируем библиотеку для работы с кодировщиками
# import category_encoders as ce 

# создаем объект OrdinalEncoder, col - имя столбца, mapping - словарь с описанием кодировки
ord_encoder = ce.OrdinalEncoder(mapping=[{
	'col': 'size',
	'mapping': {'xxs': 1, 'xs': 2, 's': 3, 
                'm': 4, 'l': 5, 'xxl': 6}
}])

clothing=clothing_itol.copy()
# применяем трансформацию к столбцу
data_bin = ord_encoder.fit_transform(clothing['size'])
# добавляем результат к исходному DataFrame
clothing = pd.concat([clothing, data_bin], axis=1)

clothing
```

### 5.9
Определите типы признаков и закодируйте их в соответствии с изученными способами

```python
list_of_dicts = [
 {'product': 'Product1', 'price': 1200, 'payment_type': 'Mastercard'},
 {'product': 'Product2', 'price': 3600, 'payment_type': 'Visa'},
 {'product': 'Product3', 'price': 7500, 'payment_type': 'Amex'}
]
df = pd.DataFrame(list_of_dicts)

df
```

```python
df_1=df.copy()

ord_encoder = ce.OrdinalEncoder()
year_col = ord_encoder.fit_transform(df_1['product'])
df_1 = pd.concat([df_1, year_col], axis=1)

encoder = ce.OneHotEncoder(cols=['payment_type'], use_cat_names=True) # указываем столбец для кодирования
type_bin = encoder.fit_transform(df_1['payment_type'])
df_1 = pd.concat([df_1, type_bin], axis=1)

df_1
```
