```python
import matplotlib.pyplot as plt


# Создание символьной переменной
x = sympy.symbols('x')

# Преобразование в функцию, пригодную для использования с библиотекой Matplotlib
y_func = sympy.lambdify(x, f, 'numpy')

# Генерация значений x от -10 до 10
x_values = np.linspace(-2, 6, 400)
# Вычисление соответствующих значений y
y_values = y_func(x_values)

# Построение графика с использованием Matplotlib
plt.plot(x_values, y_values, label=r'$-x^4+6x^3-4x^2+80$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции $y = -x^4+6x^3-4x^2+80$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

### Задание 1.4
Допустим, вы хотите произвести некоторое количество товара, которое зависит от часов работы двух ключевых сотрудников следующим образом:
$$f(x, y)=x^{2}+2 y^{2}$$
где $x$ — время работы 1-ого сотрудника, $y$ — время работы 2-ого сотрудника.

Однако вы должны оплатить этим сотрудникам ровно 20 часов работы в сумме, то есть:
$$x+y=20$$
Какое наибольшее количество товаров вы сможете произвести в таком случае?

Помните, что количество часов работы должно быть целым, поэтому, прежде чем вычислять итоговый результат, округлите часы работы до целого.

```python
x, y, w = symbols('x y w')
q=x+y-20
f=x**2+2*y**2
fq=x**2+2*y**2+w*q

print('Функция Лагранжа :\n ',fq)
fx = fq.diff(x)
print('df/dx =',fx,'= 0')
fy = fq.diff(y)
print('df/dy =',fy,'= 0')
fw = fq.diff(w)
print('df/dw =',fw,'= 0')
sols = solve([fx,fy,fw],x,y,w)
x_t=float(sols[x].round())
y_t=float(sols[y].round())
print('w :', sols[w])
print('Количество рабочих часов сотрудника "x" и y":\n',x_t,',',y_t)
print('0^2+2*20^2=800')
print('Количество производимого тавара :',f.subs({x:0,y:20}))
```
