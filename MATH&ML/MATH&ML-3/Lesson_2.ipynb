### Задание 2.2
Результат действия оператора на вектор называют образом вектора.

a) Найдите образ вектора v при действии оператора F, если
$$  \vec{v} = \left(\begin{array}{c} 1 \\ 2 \\ 3 \end{array}\right)  F = \left(\begin{array}{ccc} 1 & 0 & 0 \\ 0& 0 & 1 \\ 0 & -1 & 0 \end{array}\right)$$

```python
F=np.array([[1,0,0],[0,0,1],[0,-1,0]])
v=np.array([[1],[2],[3]])

x=F@v
print('a) новый вектор x  ',x)
print
print('Длинны векторов v={}, x={}'.format(np.linalg.norm(v),np.linalg.norm(x)))
print
v=(np.reshape(v,(1,3))[0])
x=x.T[0]
print("(v,x):",np.dot(v,x))
```

### Задание 2.3
a) Найдите образы векторов u и v при действии оператора F, если

$$ \vec{u} = \left(\begin{array}{c} 1 \\ -1 \end{array}\right) 
    \vec{v} = \left(\begin{array}{c} 1 \\ 1 \end{array}\right) 
    F = \left(\begin{array}{cc} 2 & 1 \\ 1 & 2  \end{array}\right) $$

```python
F=np.array([[2,1],[1,2]])
v=np.array([[1],[1]])
u=np.array([[1],[-1]])
print('Fu= ',F@u)
print('Fv= ',F@v)

v=(np.reshape(v,(1,2))[0])
u=u.T[0]
print("(v,x):",np.dot(v,u))
```
