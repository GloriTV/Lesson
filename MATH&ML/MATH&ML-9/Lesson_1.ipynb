```python
def f(x, y):
    return x**2+x*y-y+y**3
def grad(x, y):
    dx = 2*x+y
    dy = x+3*y**2-1
    return (dx, dy)
```

```python
x0 = (1.5, 0.5)
gamma = 0.001 # 0.1
x_cur = x0
stop_i= 10000

vals = []
coords = []
i = 0
while True:
    x_new = (x_cur[0] - gamma * grad(*x_cur)[0],
            x_cur[1] - gamma * grad(*x_cur)[1])
    if i > stop_i:
        break
    x_cur = x_new
    vals.append(f(*x_cur))
    coords.append(x_cur)
    i += 1
    print(f"iter={i}; x=({x_cur[0]:.6f}, {x_cur[1]:.6f});"
          f" f(x)={f(*x_cur):.6f}; grad f(x)=({grad(*x_cur)[0]:.6f}, {grad(*x_cur)[1]:.6f})")
    
print(f"Ответ : {x_cur[1]:.2f}")
```
